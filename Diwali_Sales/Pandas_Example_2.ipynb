```python
df.drop(['Status','unnamed1'], axis = 1, inplace = True)
```

The `inplace` argument in Python functions specifies whether the function modifies the original object or creates a copy.

When `inplace` is set to `True`, the function modifies the original object and returns None. This can be more efficient than creating a new copy of the object, especially for large datasets. However, it also means that the original object is modified, which can be undesirable in some cases.

When `inplace` is set to `False` (the default), the function creates a copy of the object and returns the copy. This ensures that the original object is not modified, but it can be less efficient.

---

```python
# You can use dfs = df.dropna but it will create a copy of your data
df.dropna(inplace = True)
```

The `df.dropna(inplace = True)` method in Pandas is used to remove any rows that contain missing values `(NaN)` from a DataFrame. It modifies the original DataFrame inplace, meaning that the changes are permanent.

---
The main difference between `df.dropna(inplace = True)` and dfs = df.dropna is that `df.dropna(inplace = True)` modifies the original DataFrame, while `dfs = df.dropna` creates a copy of the DataFrame with the dropped rows.

`df.dropna(inplace = True)` :
* Modifies the original DataFrame inplace.
* Returns None.

`dfs = df.dropna` :
* Creates a copy of the DataFrame with the dropped rows.
* Returns the new DataFrame.

There are two other methods you can use to rename columns in a Pandas DataFrame without using inplace=True:

1. Using the `.assign()` method:

`df = df.assign(Married=df['Marital_Status'])`
This method creates a new DataFrame with the new column name, while leaving the original DataFrame unchanged.

2. Using dictionary assignment:

`df['Married'] = df['Marital_Status']`

This method directly assigns the values from the `Marital_Status` column to a new column named `Married`.

```python
# Analysis on the bases of age group and gender, sort by count

ay = sns.countplot(data = df, x = 'Age Group', hue = 'Gender', order=df.groupby('Age Group')['Gender'].count().sort_values(ascending=False).index)

for bars in ay.containers:
  ay.bar_label(bars)
```

## State

```python
# total number of orders from top 10 states

# Create a dataframe group by State
sales_state = df.groupby(['State'], as_index = False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

# Set the figure size
sns.set(rc={'figure.figsize':(20,5)})

# Create the plot
sns.barplot(data=sales_state, x = 'State', y = 'Orders', palette="hls", width =0.5)
```

Top states are UP, Maharashtra, Karnataka

```python
# total number of orders from top 10 states

# Create a dataframe group by State
sales_state = df.groupby(['State'], as_index = False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

# Set the figure size
sns.set(rc={'figure.figsize':(20,5)})

# Create the plot
sns.barplot(data=sales_state, x = 'State', y = 'Amount', palette="hls", width =0.5)
```

## Marital Status

```python
# Create the countplot
ay = sns.countplot(data=df, x="Marital_Status", width=0.4, palette="hls")

# Set the figure size
sns.set(rc={"figure.figsize": (5, 4)})

# Set the x-axis tick labels
ay.set_xticklabels(["Married", "Non-Married"])

# Hide the y-axis
ay.yaxis.set_visible(False)

# Add data labels
for bars in ay.containers:
    ay.bar_label(bars)

# Display the plot
plt.show()
```

Married Buyer are more then non-married

```python
# Create a Data Frame for Marital_Status, Gender wrt Amount Spent
sales_status = df.groupby(['Marital_Status','Gender'],as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending= False)

# Create a bar plot
ay = sns.barplot(data = sales_status, x = 'Marital_Status', y= 'Amount', hue = 'Gender')

# Set the x-axis tick labels
ay.set_xticklabels(["Married", "Non-Married"])
```

Femal married buyer are more them married male buyer

## Occupation

```python
# Describe the size of plot
sns.set(rc={'figure.figsize':(20,5)})

# Create a countplot, sorting by the number of occurrences
ax = sns.countplot(data = df, x= 'Occupation', order=df['Occupation'].value_counts().index, palette="hls")

#this to rotate the x axis lable
# plt.xticks(rotation=90, multialignment='left')

for bars in ax.containers:
  ax.bar_label(bars)
```

Top 3 sectors are IT Sector, Healthcare, Aviation

```python
sns.set(rc={'figure.figsize':(20,5)})

# Sort the DataFrame by 'Occupation' based on value counts
df_sorted = df.groupby('Occupation').size().reset_index(name='count').sort_values(by='count', ascending=False)

# Create a countplot, sorting by the number of occurrences
ax = sns.countplot(data=df, x='Occupation', order=df_sorted['Occupation'], palette="hls", width=0.5)

# Wrap the x-axis labels
wrapper = textwrap.TextWrapper(width=10, break_long_words=True)

# Wrap all the labels
wrapped_labels = [wrapper.fill(label) for label in df_sorted['Occupation']]

# Set the x-axis tick labels to the wrapped labels
ax.set_xticklabels(wrapped_labels)

# Show the plot
plt.show()
```

## Product Category

```python
# Sort the DataFrame by 'Occupation' based on value counts
df_sorted = df.groupby('Product_Category').size().reset_index(name='count').sort_values(by='count', ascending=False)

# Create a countplot, sorting by the number of occurrences
ax = sns.countplot(data=df, x='Product_Category', order=df_sorted['Product_Category'], palette="hls", width=0.5)

# Wrap the x-axis labels
wrapper = textwrap.TextWrapper(width=10, break_long_words=True)

# Wrap all the labels
wrapped_labels = [wrapper.fill(label) for label in df_sorted['Product_Category']]

# Set the x-axis tick labels to the wrapped labels
ax.set_xticklabels(wrapped_labels)

for bars in ax.containers:
  ax.bar_label(bars)

# Show the plot
plt.show()
```

Top 3 Category are Cloths, Food and E&G

```python
# Sort the DataFrame by 'Occupation' based on value counts
df_sorted = df.groupby(['Product_Category'],as_index = False)['Amount'].sum().sort_values(by='Amount', ascending=False)

# Create a countplot, sorting by the number of occurrences
ax = sns.barplot(data = df_sorted, x='Product_Category', y='Amount', order=df_sorted['Product_Category'], palette="hls", width=0.5)

# Wrap the x-axis labels
wrapper = textwrap.TextWrapper(width=10, break_long_words=True)

# Wrap all the labels
wrapped_labels = [wrapper.fill(label) for label in df_sorted['Product_Category']]

# Set the x-axis tick labels to the wrapped labels
ax.set_xticklabels(wrapped_labels)

# Show the plot
plt.show()
```
