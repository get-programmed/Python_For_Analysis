```python
# Import the needed library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

*   pandas: a library used for data manipulation and analysis.
*   numpy: a library used for scientific computing and array operations.
*   matplotlib.pyplot: a library used for creating visualizations.
---
---

```python
df = pd.read_csv(r"https://raw.githubusercontent.com/rishabhnmishra/Python_Diwali_Sales_Analysis/main/Diwali%20Sales%20Data.csv",header=0,encoding='latin1')
```

1. `pd.read_csv`: This function from the Pandas library is used to read a CSV file into a DataFrame.

2. r"https://raw.githubusercontent.com/rishabhnmishra/Python_Diwali_Sales_Analysis/main/Diwali%20Sales%20Data.csv": This is the URL of the CSV file that you want to read. The r before the string indicates that it is a raw string, meaning that backslashes are not treated as escape characters.

3. header=0: This argument specifies that the first row of the CSV file contains the column names.

4. encoding='latin1': This argument specifies the character encoding of the CSV file. In this case, it is set to latin1, which is commonly used for data from Western Europe.

---
---


*   A DataFrame is a two-dimensional data structure that stores data in rows and columns(Tabular data).
*   A Data Series is a one-dimensional data structure that stores data in a sequence.

```python
#saves the DataFrame df to a CSV file named Diwali_sales.csv. The file will be saved in the current working directory
df.to_csv('Diwali_sales')
```

* `columns`: A list of specific columns to include in the CSV file.
* `encoding`: The character encoding to use when saving the CSV file.

```python
df.iloc[3]
```

The `df.iloc[3]` expression in Pandas selects the fourth row of the DataFrame df.


*   `df`: This refers to the DataFrame object itself.
*   `iloc`: This is a method of the DataFrame object that allows you to select data using numerical indices.
*   `[3]`: This is the numerical index of the row you want to select. In Python, indexing starts at 0, so the fourth row has an index of 3.

```python
df['Orders'] = df['Orders'].astype(str)
df['Orders']
```

*   `df['Orders'] = df['Orders'].astype(str)`: This line converts the data type of the Orders column from its original type (which could be integer, float, etc.) to string using the astype(str) method.

```python
# This selects rows from index 6 (inclusive) to index 8 (exclusive).
df['Orders'][6:9]
```

```python
df[['Orders','Age']]
```

The code `df[['Orders','Age']]` will select the columns named "Orders" and "Age" from the DataFrame df and return a new DataFrame containing only those two columns.

Here's a breakdown of what the code does:

1. Accessing the DataFrame:
* The variable df is assumed to be a reference to a pandas DataFrame object.
---
2. Column Selection:
* `[]`: This is the indexing operator in pandas. It allows you to select specific rows or columns from a DataFrame.
* `['Orders','Age']`: This is a list containing the names of the two columns you want to select.
---
3. Result:
* The code returns a new DataFrame that contains only the "Orders" and "Age" columns from the original DataFrame df.

---

```python
# .loc[0]: This selects the row at index 0 from the resulting DataFrame.
df[['Orders','Age']].loc[0]
```

```python
# .squeeze(): This method removes any single-dimensional elements from the DataFrame, resulting in a Series object.
df[['Orders','Age']].loc[0].squeeze()
```

```python
df.iat[0,2]
```

The df.iat[0,2] expression in Pandas selects the element at row 0 and column 2 in the DataFrame df.

*  `df`: This refers to the DataFrame object itself.
*   `iat`: This is a method of the DataFrame object that allows you to select a single element using its numerical index.
*   `[0,2]`: This is a tuple of numerical indices. The first element (0) represents the row index, and the second element (2) represents the column index.
