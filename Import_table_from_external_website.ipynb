## **Import table from external ebsite using python**

```python
import pandas as pd
```

```python
url = 'https://en.wikipedia.org/wiki/Production_car_speed_record'

dfs = pd.read_html(url)
```

* The `read_html()` function requires a parser to handle the HTML content.
* By default, it uses the `bs4` parser, which requires the `BeautifulSoup` library.
* Since `BeautifulSoup` is not a mandatory dependency for Jupyter notebooks running on Linux, it's recommended to use the `html5lib` parser instead.

```python
df = dfs[1]
```

The code `df = dfs[1]` assigns the second element of the list dfs to the variable df.

Here's what the code does:

* `dfs`: This is a list that contains multiple DataFrames.
* `[1]`: This is the index of the second element in the list.
* `df = dfs[1]`: This assigns the second element (which is a DataFrame) to the variable df.

```python
dfs[1]
```

```python
styled_table = df.iloc[:3].style.set_table_attributes("class='mytable'").set_properties(**{'text-align': 'left'})

display(styled_table)
```

The code `styled_table = df.iloc[:3].style.set_table_attributes("class='mytable'").set_properties(**{'text-align': 'left'})` creates a styled DataFrame with the following characteristics:

* Only the first 3 rows of the DataFrame `(df.iloc[:3])` are included.
* The CSS class `mytable` is applied to the table element.
* The text alignment of all cells is set to left.
Here's a breakdown of what the code does:

* `df.iloc[:3]`: This selects the first 3 rows of the DataFrame using the iloc attribute.
* `.style`: This accesses the Pandas styling object associated with the selected rows.
* `.set_table_attributes("class='mytable'")`: This sets the class attribute of the table element to mytable.
* `.set_properties({'text-align': 'left'})`**: This sets the text alignment of all cells in the table to left.
* `display(styled_table)`: This displays the styled DataFrame as an HTML table.
When you run this code, you will see a table with the following properties:

* Only the first 3 rows of the DataFrame are displayed.
* The table has a CSS class called `mytable`.
* The text in all cells is aligned to the left.
